#### a) What are the steps of `kmeans`?
**Hint**: There are 4 steps/builder functions that you'll need.

**1. Assign each point to a cluster N at random.  
2. Calculate the mean position of each cluster using the previous assignments.  
3. Loop through the points - assign each point of the cluster to whose center it is closest.  
4. Repeat this process until the centers stop moving around.**

#### b) Create the builder function for step 1.

```r
initialize_clusters <- function(data, k){
    n <- nrow(data)
    cluster_assignments <- sample(1:k, n, replace = TRUE)
    return(cluster_assignments)
}
```

#### c) Create the builder function for step 2.

```r
compute_centroids <- function(data, clusters, k){
    centroids <- matrix(NA, nrow = k, ncol = ncol(data))
    colnames(centroids) <- colnames(data)
    for (i in 1:k){
        cluster_points <- data[clusters == i, , drop = FALSE]
        centroids[i, ] <- colMeans(cluster_points)
    }
    return(centroids)
}
```

#### d) Create the builder function for step 3.
*Hint*: There are two ways to do this part - one is significantly more efficient than the other. You can do either.  

```r
assign_clusters <- function(data, centroids){
  data <- as.matrix(data)
  centroids <- as.matrix(centroids)
  storage.mode(data) <- "numeric"
  storage.mode(centroids) <- "numeric"
  
  distances <- outer(rowSums(data^2), rowSums(centroids^2), "+") - 2 * data %*% t(centroids)
  new_clusters <- max.col(-distances)
  return(new_clusters)
}
```

#### e) Create the builder function for step 4.

```r
kmeans_algorithm <- function(data, k, max_iter = 100){
    clusters <- initialize_clusters(data, k)
    for (i in 1:max_iter){
        centroids <- compute_centroids(data, clusters, k)
        new_clusters <- assign_clusters(data, centroids)
        if (all(new_clusters == clusters)){
            cat("Converged after", i, "iterations.\n")
            break
        }
        cluster <- new_clusters
    }
    return(list(clusters = clusters, centroids = centroids))
}
```

#### f) Combine them all into your own `kmeans` function.

```r
# step 1
initialize_clusters <- function(data, k){
    n <- nrow(data)
    cluster_assignments <- sample(1:k, n, replace = TRUE)
    return(cluster_assignments)
}
# step 2
compute_centroids <- function(data, clusters, k){
    centroids <- matrix(NA, nrow = k, ncol = ncol(data))
    colnames(centroids) <- colnames(data)
    for (i in 1:k){
        cluster_points <- data[clusters == i, , drop = FALSE]
        centroids[i, ] <- colMeans(cluster_points)
    }
    return(centroids)
}
# step 3
assign_clusters <- function(data, centroids){
  data <- as.matrix(data)
  centroids <- as.matrix(centroids)
  storage.mode(data) <- "numeric"
  storage.mode(centroids) <- "numeric"
  
  distances <- outer(rowSums(data^2), rowSums(centroids^2), "+") - 2 * data %*% t(centroids)
  new_clusters <- max.col(-distances)
  return(new_clusters)
}

# step 4
kmeans_algorithm <- function(data, k, max_iter = 100){
    clusters <- initialize_clusters(data, k)
    for (i in 1:max_iter){
        centroids <- compute_centroids(data, clusters, k)
        new_clusters <- assign_clusters(data, centroids)
        if (all(new_clusters == clusters)){
            cat("Converged after", i, "iterations.\n")
            break
        }
        clusters <- new_clusters
    }
    return(list(clusters = clusters, centroids = centroids))
}
```

## Question 3
#### a) Explain the process of using a for loop to assign clusters for kmeans.

**In K-Means, using a for loop to assign clusters means iterating through each data point one by one, calculating its distance from every cluster centroid, and assigning it to the cluster with the smallest distance. This approach is straightforward to follow, but it requires nested loops (one loop for data points and another one for centroids), which makes it slow for large datasets.**

#### b) Explain the process of vectorizing the code to assign clusters for kmeans.

**Vectorizing the code replaces explicit loops with efficient matrix operations. Instead of calculating distances for each point separately, all distances between data points and centroids are computed simultaneously using vectorized matrix algebra functions like %*%. Then, functions such as max.col(-distances) are used to assign each point to the nearest centroid in one step. This approach leverages optimized, low-level operations in R for speed.**

#### c) State which (for loops or vectorizing) is more efficient and why.

**Vectorizing is more efficient than using for loops because it allows operations to be performed on entire arrays or matrices at once instead of processing each element individually. This takes advantage of low-level, optimized compile languages within R, reducing the amount of interpreted R code that needs to run and minimizing memory overhead. As a result, vectorized code executes faster and is generally more concise and readable.**

## Question 4
#### When does `kmeans` fail? What assumption does `kmeans` use that causes it to fail in this situation?

**K-Means fails when the data does not form clear, round, and evenly sized clusters. It has trouble when clusters have different shapes, sizes, or densities, or when they overlap. This happens because K-Means assumes that each cluster is roughly spherical and that the distance to the cluster center (centroid) is a good way to measure similarity. When this assumption is not true, K-Means cannot separate the data correctly.**

## Question 5
#### What assumption do Guassian mixture models make?

**Gaussian Mixture Models assume that the data is generated from a mixture of several Gaussian distributions, where each cluster follows its own Gaussian shape with its own mean and covariance. This means GMM assumes that data points within a cluster are normally distributed around the cluster center, and the overall data distribution is a combination of these individual Gaussian components.**

## Question 6
#### What assumption does spectral clustering make? Why does this help us?

**Spectral clustering assumes that data points that belong to the same cluster are strongly connected or have a high similarity, and points from different clusters are weakly connected. This assumption is based on the idea that the data can be represented as a graph, where edges represent similarities between points. By analyzing the eigenvalues and eigenvectors of this graph’s Laplacian matrix, spectral clustering can find clusters of almost any shape, not just spherical ones. This helps because it allows the algorithm to detect complex, non-linear cluster boundaries that methods like K-Means would fail to capture.**

## Question 7
#### Define the gap statistic method. What do we use it for?

**The gap statistic method is a technique used to determine the optimal number of cluster *k* in a dataset. It works by comparing the total within-cluster variation for different values of k with the expected variation under a reference null distribution. The idea is to find the k where the gap between the observed clustering performance and the random reference is the largest. In other words, we use the gap statistic to identify the number of clusters that best separates the data compared to random noise.**

## Question 9

Which version of the counter function works? What is the difference in the way R processes the two functions?

```r

counter1 <- function(start, step){
    val <- start;
    function(){
        old_val <- val;
        val <- val + step;
        old_val;
    }
}
counter_from_1 <- counter1(1,1);
counter_from_1()
counter_from_1()

counter2 <- function(start, step){
    val <- start;
    function(){
        old_val <- val;
        val <<- val + step;
        old_val;
    }
}
counter_from_1 <- counter2(1,1);
counter_from_1()
counter_from_1()
```

We can see that the only difference between function counter1 and function counter2 is that: one using "<-" to pass the value, another using "<<-"

The difference between those two operators is that "<-" is used to assign a value to the variable in the current environment, and "<<-" is used to assign a value to the variable in the first parent environment where it already exists. If it doesn't exist in any parent, it creates a new variable in the global environment.

For counter1, val will always be 1, because in this step, "val <- val + step;", it always assigns 1 to val. 

## Question 10

a)  Use read_csv the cars.csv and store it in a data frame.
b)  Then, group the data frame by Make, get averages across the numeric variables, and then sort by Volume in descending order. Hint: Use summarise(across(c(<variable list>), mean)) to get the averages.

```r
library(readr)
cars_df <- read_csv("cars.csv")
cars_df
```

```r
library(dplyr)
cars_result <- cars_df %>%
group_by(Make) %>%
summarise(across(c(Volume, Weight, CO2), mean)) %>%
arrange(desc(Volume))
cars_result
```

## Question 11

Make a function that returns the Fibonacci sequence, then call it 7 times to return the first 7 values of the sequence. Use the correct counter function from question 9 for inspiration.

```r
fibonacci <- function(){
    x <- 0
    y <- 1
    function(){
        temp <- x + y
        x <<- y
        y <<- temp
        x
    }
}
```

```r
result <- fibonacci()

for (i in 1:7){
print(result())
}
```

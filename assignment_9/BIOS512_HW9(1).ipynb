## Question 2 
#### a) What shape does a perfect classifier look like on an ROC curve? What about a bad classifier?
**A perfect classifier’s ROC curve hugs the **top-left corner** (AUC = 1), showing both high true positive rate and low false positive rate.  
A bad classifier’s ROC curve lies **close to the diagonal line**, meaning it performs no better than random guessing.**
#### b) Think about the formula for an F1 score. What does it mean when the F1 score is close to 1? Close to 0?
**When the **F1 score is close to 1**, the model has **both high precision and high recall**, indicating excellent performance.  
When the **F1 score is close to 0**, the model performs **poorly**, with low precision, recall, or both.**

It is important to train and then test a model to ensure that it can generalize to new, unseen data rather than just memorizing the training data.

## Question 5
This question runs through a linear regression example. We want to predict median house value based on the other variables.
#### a) First, load the `housing.csv` data set. Look at the data in some useful way. Why is linear regression appropriate here?

```r
housing <- read.csv("housing.csv", stringsAsFactors = FALSE)

head(housing)
str(housing)
summary(housing)

na_counts <- colSums(is.na(housing))
na_counts

target <- "median_house_value"

hist(housing[[target]], main="median_house_value", xlab=target)

if ("median_income" %in% names(housing)) {
  plot(housing$median_income, housing[[target]],
       xlab="median_income", ylab=target)
}

num_cols <- names(housing)[sapply(housing, is.numeric)]
num_cols <- setdiff(num_cols, target)
cor_with_target <- sapply(num_cols, function(col)
  suppressWarnings(cor(housing[[col]], housing[[target]], use="complete.obs")))
sort(round(cor_with_target, 3), decreasing = TRUE)
```

Linear regression is appropriate here because the target variable, `median_house_value`, is continuous.  
From the summary statistics and histogram, the data show a roughly linear relationship between `median_income` and `median_house_value` (correlation ≈ 0.69), suggesting that as income increases, house value tends to increase.  
Since the predictors are numeric and the relationship appears approximately linear, linear regression is a reasonable starting model for this dataset.

#### b) Scale data and split it 75/25 training/testing. Set seed = 123.

```r
set.seed(123)

# Scale numeric features
housing_scaled <- as.data.frame(scale(housing))

# Split into training (75%) and testing (25%)
n <- nrow(housing_scaled)
train_index <- sample(1:n, size = 0.75 * n)

train <- housing_scaled[train_index, ]
test  <- housing_scaled[-train_index, ]
```

#### c) Fit the model.

```r
# Fit linear regression model
model <- lm(median_house_value ~ ., data = train)

summary(model)
```

#### d) Make predictions on test data and show them in an actual vs. predicted plot.

```r
pred <- predict(model, newdata = test)
actual <- test$median_house_value

plot(actual, pred,
     xlab = "Actual (standardized)", ylab = "Predicted (standardized)",
     main = "Actual vs. Predicted")
abline(0, 1)

# (optional) RMSE
sqrt(mean((pred - actual)^2))
```

#### e) Make a residuals plot.

```r
residuals <- actual - pred

plot(pred, residuals,
     xlab = "Predicted (standardized)", 
     ylab = "Residuals",
     main = "Residuals Plot")
abline(h = 0, col = "red")
```

## Question 6
This question runs through a logistic regression example. We want to predict diabetes diagnosis based on the other variables. 
#### a) First, load the `diabetes.csv` data set. Look at the data in some useful way. Why is logistic regression appropriate here?

```r
# Load dataset
diabetes <- read.csv("diabetes.csv", stringsAsFactors = FALSE)

# Basic checks
head(diabetes)
str(diabetes)
summary(diabetes)

# Check class balance
table(diabetes$Outcome)

cor(diabetes[sapply(diabetes, is.numeric)])
```

Logistic regression is appropriate here because the response variable `Outcome` is **binary** (0 = no diabetes, 1 = diabetes).  
The summary statistics show that the predictors like `Glucose`, `BMI`, and `Age` are continuous numeric variables that can influence the probability of having diabetes.  
The correlation matrix also shows that `Glucose` (r ≈ 0.47) and `BMI` (r ≈ 0.29) have positive relationships with `Outcome`, suggesting that higher values of these features increase the likelihood of diabetes — which fits the **probabilistic nature of logistic regression**.

#### b) Scale data and split it 75/25 training/testing. Set seed = 123.

```r
set.seed(123)

# Scale numeric predictors
diabetes_scaled <- as.data.frame(scale(diabetes[, -9])) 
diabetes_scaled$Outcome <- diabetes$Outcome           

# Split 75/25
n <- nrow(diabetes_scaled)
train_index <- sample(1:n, size = 0.75 * n)

train <- diabetes_scaled[train_index, ]
test  <- diabetes_scaled[-train_index, ]
```

#### c) Fit the model.

```r
# Fit logistic regression model
model_log <- glm(Outcome ~ ., data = train, family = binomial)

summary(model_log)
```

#### d) Make predictions on test data. Print a table with the number of true positives, false positives, true negatives, false negatives, and accuracy. 

```r
# Make predictions on test data
prob <- predict(model_log, newdata = test, type = "response")

# Convert probabilities to binary predictions
pred <- ifelse(prob > 0.5, 1, 0)

# Confusion matrix
conf_mat <- table(Predicted = pred, Actual = test$Outcome)
conf_mat

# Extract counts
TP <- conf_mat["1", "1"]
FP <- conf_mat["1", "0"]
TN <- conf_mat["0", "0"]
FN <- conf_mat["0", "1"]

# Calculate accuracy
accuracy <- (TP + TN) / sum(conf_mat)

cat("True Positives:", TP, "\n")
cat("False Positives:", FP, "\n")
cat("True Negatives:", TN, "\n")
cat("False Negatives:", FN, "\n")
cat("Accuracy:", round(accuracy, 4), "\n")
```

#### e) Fit a LASSO-regularized logistic regression model. Again, set seed = 123. Which variables are the most important (which ones don't go to zero)? How does the LASSO model affect the accuracy?

```r
install.packages("glmnet")
library(glmnet)

# Set seed
set.seed(123)

# Prepare data
x_train <- as.matrix(train[, -9])
y_train <- train$Outcome
x_test <- as.matrix(test[, -9])
y_test <- test$Outcome

# Fit LASSO model
lasso_model <- cv.glmnet(x_train, y_train, family = "binomial", alpha = 1)

# View best lambda
lasso_model$lambda.min

# Coefficients at best lambda
coef(lasso_model, s = "lambda.min")

# Predictions
prob_lasso <- predict(lasso_model, newx = x_test, s = "lambda.min", type = "response")
pred_lasso <- ifelse(prob_lasso > 0.5, 1, 0)

# Confusion matrix
conf_mat_lasso <- table(Predicted = pred_lasso, Actual = y_test)
conf_mat_lasso

# Accuracy
accuracy_lasso <- sum(diag(conf_mat_lasso)) / sum(conf_mat_lasso)
accuracy_lasso
```

The LASSO-regularized logistic regression kept the coefficients for **Pregnancies, Glucose, BloodPressure, BMI, DiabetesPedigreeFunction,** and **Age** as nonzero, meaning these are the most important predictors of diabetes diagnosis.  
Variables like **SkinThickness** and **Insulin** were shrunk close to zero, indicating low predictive contribution.  
The LASSO model achieved an accuracy of **0.766**, which is almost the same as the standard logistic regression model (0.771).  
This shows that LASSO maintains predictive performance while simplifying the model and reducing overfitting.

#### f) Make a plot of actual vs. predicted values for the LASSO model.

```r
plot(prob_lasso, y_test,
     xlab = "Predicted Probability (LASSO)",
     ylab = "Actual Outcome",
     main = "Actual vs Predicted (LASSO Logistic Regression)",
     col = "blue", pch = 19)
abline(h = 0.5, col = "red", lwd = 2)
```

The plot shows the relationship between the predicted probabilities and the actual diabetes outcomes for the LASSO logistic regression model.  
Points near 0 on the y-axis correspond to non-diabetic cases, while points near 1 correspond to diabetic cases.  
The red horizontal line at 0.5 represents the classification threshold — predictions above this are classified as positive (1).  
Most predicted probabilities cluster near 0 or 1, indicating that the model confidently distinguishes between the two classes, consistent with an accuracy of about 0.77.
